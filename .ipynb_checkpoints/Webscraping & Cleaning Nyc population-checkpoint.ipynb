```python
# new york city population over time fro wiki
import requests
from bs4 import BeautifulSoup
```

```python
# check website response
pop = requests.get('https://en.wikipedia.org/wiki/New_York_City')
pop
```

```python
# get text works when you index it to a paritcular tag
soup = BeautifulSoup(pop.content,'html.parser')
soup.find_all('tr')[0].get_text()
```

```python
# Exlpored html, found the table and data within.
Historical_table = soup.find(class_ ="toccolours")
table_body = Historical_table.find('tbody')
table_item = table_body.find('th')
```

```python
table_item
```

```python
year = table_item.text
```

```python
year
```

#### After a while, I thought why not try CSS selectors, since I know the tags that house the data and in what order .

```python
# The data for years are  stored in a th tag, will return the same output as our find methods
item = Historical_table.select("body tr th")
```

```python
# the  first 3 values are the column names, which I dont want
# I can add them later as I make a dataframe.
periods = [pt.get_text() for pt in item[3::]]
len(periods)
```

```python
# To get the rest of the data,we must swith to the td tag
item_v = Historical_table.select("body tr td")
```

```python
popu = [pt.get_text() for pt in item_v[0:-1:2]]
len(popu)
```

```python
pctc= [pt.get_text() for pt in item_v[1::2]]
len(pctc)
```

# Turning our Data into a Dataframe

```python
import pandas as pd
```

```python
Headers =['Years', "Population", "Percent_Change"]
```

```python
Nyc_historical = pd.DataFrame({
    'Years':periods,
    "Population": popu,
    "Percent_Change":pctc
    
})
Nyc_historical
```

```python
# Need to convert these objects in to integers.
Nyc_historical.info()
```

# Problems encountered 
1. ~~With populations,I cant convert it into a numeric(int), due to the commas~
2. ~~need to seperate the  numeric in percent Change from the symbols andd probabluy convert it back into a decimal (float)~~

```python
Nyc_historical['Population'] = Nyc_historical['Population'].str.replace(",","").astype(int)
Nyc_historical['Percent_Change'] = Nyc_historical['Percent_Change'].str.extract('(\d+\.\d)')
```

### More problems spoted, more cleaning :
     Another problem that arosed was the '---' at index 0 acts as a NaN, so I cant convert the percent change column,
     without filling in that cell  with a 0. Which I rather do than droping that row of data. as this column
     is just  reporting (aggregation) the change over the given time, since theres no data before  the first index
     its make senses there would be nothing in that column. Though changing type to float works too, with that  value error          caused by of that Nan

```python
# optional
Nyc_historical['Percent_Change'] = Nyc_historical['Percent_Change'].fillna(0)


# also optional 
# Saves me the trouble later , but in all honestly an  unnessary step because of the column header 
# that reminds me that those values in percent changes are percentages not whole numbers
decim = []
for x in  Nyc_historical['Percent_Change']:
    dec = x/100
    de = round(dec,3)
    decim.append(de)
Nyc_historical['dec_Change']=decim
```

```python
Nyc_historical['Percent_Change'] = Nyc_historical['Percent_Change'].astype(float)
# since its only the years and not a actual date, we convert it to a integer
Nyc_historical['Years'] = Nyc_historical['Years'].astype(int)

Nyc_historical
```

```python
# luckly pandas has its own save to function
#index is false as to_csv by default saves everying in the dataframe , even the temporary index
Nyc_historical.to_csv('Nyc_historical_pop.csv', sep ='\t',index=False)

```
